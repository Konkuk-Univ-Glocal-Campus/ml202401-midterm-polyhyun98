```python
#필요한 라이브러리 import
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import seaborn as sns


```

```python
#불러온 csv 파일  인코딩 하여 깨진 파일 복구
df = pd.read_csv('경찰청_범죄 발생 시간대 및 요일_20191231.csv', encoding = 'euc-kr')
```

```python
#데이터 프레임 확인
df
```

```python
# 클러스터링을 위한 입력 데이터 생성
X = df.drop(columns=['범죄대분류', '범죄중분류'])
```

```python
# 데이터 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# K-means 클러스터링 적용
wcss = []
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
# 엘보우 방법을 사용하여 최적의 클러스터 수 결정
plt.plot(range(2, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
# 클러스터링 결과 시각화
best_k = silhouette_scores.index(max(silhouette_scores)) + 2
kmeans_best = KMeans(n_clusters=best_k, random_state=42)
kmeans_best.fit(X_scaled)
df['Cluster'] = kmeans_best.labels_
```

```python
# 클러스터별 분류 시각화
plt.figure(figsize=(10, 6))
for cluster in range(best_k):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['범죄대분류'], cluster_data['범죄중분류'], label=f'Cluster {cluster+1}')
plt.xlabel('범죄대분류')
plt.ylabel('범죄중분류')
plt.title('범죄 분류 클러스터링 결과')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 클러스터 분석
cluster_centers = scaler.inverse_transform(kmeans_best.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=X.columns)
print("각 클러스터의 중심점:")
print(cluster_centers_df)

#클러스터 1(주간 평온 클러스터): 범죄 발생량이 일반적으로 낮은 편이며, 오전 시간대에 증가하는 경향을 보임. 특히, 12시부터 오후 6시까지의 시간대에 발생량이 높은 편임. 또한, 월요일부터 토요일까지 거의 일정한 수준으로 발생함을 알 수 있다.
#클러스터 2(야간 활발 클러스터): 범죄 발생량이 클러스터 1에 비해 상당히 높은 편. 특히, 저녁 시간대부터 새벽까지의 시간대에 발생량이 매우 높다. 미상인 경우의 범죄 발생량도 상당히 높으며, 월요일과 금요일에 가장 많이 발생하는 경향을 볼 수 있다.
```

```python
# 평가
print("실루엣 점수:", max(silhouette_scores))

#높은 실루엣 점수를 가진 클러스터링 결과는 범죄 발생에 대한 패턴을 파악하는 데 유용하며, 예방 및 대응 전략 수립에 도움이 될 수 있다.
```
